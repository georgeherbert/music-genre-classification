```python
from matplotlib import pyplot as plt
from matplotlib import rc

import pandas as pd

import math

rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

plt.rcParams['figure.dpi']= 300
```

```python
accuracy_train = pd.read_csv("training_curves/accuracy_train.csv")
accuracy_val = pd.read_csv("training_curves/accuracy_val.csv")
loss_train = pd.read_csv("training_curves/loss_train.csv")
loss_val = pd.read_csv("training_curves/loss_val.csv")
```

```python
plt.figure(figsize=(5, 2.5))
plt.plot(accuracy_train["Step"], accuracy_train["Value"], label="Train")
plt.plot(accuracy_val["Step"], accuracy_val["Value"], label="Validation")
plt.xlabel("Batches")
plt.ylabel("Accuracy (\%)")
plt.xlim(0, accuracy_train["Step"].max())
plt.ylim(0, 100)
plt.legend()
plt.grid(linestyle=":")
plt.savefig("report/images/accuracy.png", bbox_inches='tight', pad_inches=0.0)
plt.show()

plt.figure(figsize=(5, 2.5))
plt.plot(loss_train["Step"], loss_train["Value"], label="Train")
plt.plot(loss_val["Step"], loss_val["Value"], label="Validation")
plt.xlabel("Batches")
plt.ylabel("Loss")
plt.xlim(0, loss_train["Step"].max())
plt.ylim(
    loss_train["Value"].min(),
    10 ** math.ceil(
        math.log10(
            max(
                loss_train["Value"].max(),
                loss_val["Value"].max()
            )
        )
    )
)
plt.yscale("log")
plt.legend()
plt.grid(linestyle=":")
plt.savefig("report/images/loss.png", bbox_inches='tight', pad_inches=0.0)
plt.show()
```
